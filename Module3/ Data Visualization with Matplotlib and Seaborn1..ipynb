```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.patches import Rectangle

# Read the CSV file
df = pd.read_csv('shipment_visualization.csv')

# Set up the plotting style
plt.style.use('default')
sns.set_palette("husl")
```

# 1. LINE PLOT - Shipping Times Trend (index-based)

```python
fig = plt.figure(figsize=(12, 5))
plt.plot(df.index, df['shipping_times'], marker='o', markersize=3, linewidth=1, alpha=0.7)
plt.title('Shipping Times Trend Over Records', fontsize=14, fontweight='bold')
plt.xlabel('Record Index')
plt.ylabel('Shipping Time (days)')
plt.grid(True, alpha=0.3)

plt.tight_layout(pad=3.0)
plt.show()
```

# 📈 Line Plot – Shipping Times Trend Over Records

## 📊 Description:

### This line plot displays "Shipping Time (days)" against "Record Index" (from 0 to 300). The pink line shows the shipping time for each individual record, highlighting its fluctuation over time.
## 🔍 Insight:

### Shipping times are highly volatile, frequently fluctuating between 1 and 9 days. While many shipments complete within 2-4 days, there are numerous unpredictable spikes to much longer durations, indicating inconsistency in delivery performance.

## 📖 Story:

### Our shipping times are wildly inconsistent. While many deliveries are quick, sudden delays up to 9 days are common. This high variability highlights an urgent need to stabilize our delivery performance.

# 2. SCATTER PLOT - Distance vs Shipping Time

```python
fig = plt.figure(figsize=(10, 5))
colors = ['red', 'blue', 'green', 'orange', 'purple']
locations = df['Location'].unique()
for i, location in enumerate(locations):
    location_data = df[df['Location'] == location]
    plt.scatter(location_data['distance(km)'], location_data['shipping_times'], 
               alpha=0.6, label=location, s=30, color=colors[i])
plt.title('Distance vs Shipping Time by Location', fontsize=14, fontweight='bold')
plt.xlabel('Distance (km)')
plt.ylabel('Shipping Time (days)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout(pad=3.0)
plt.show()
```

# 📊 Scatter Plot – Distance vs. Shipping Time by Location
## 📊 Description:
### This scatter plot displays the relationship between shipping distance (in kilometers) and delivery time (in days) across five Indian cities: Mumbai (red), Chennai (blue), Kolkata (green), Delhi (orange), and Bangalore (purple). The data points are clustered in distinct distance bands, suggesting shipments originate from specific hub locations. The plot shows distances ranging from near-zero to approximately 1,600 km, with shipping times varying from 1 to 9 days.
## 🔍 Insight:
### The data reveals a complex relationship where distance alone doesn't determine shipping time. Short-distance shipments (0-400 km) consistently deliver in 1-3 days across all cities, showing efficient regional logistics. However, medium to long-distance shipments (800-1,600 km) show dramatic performance variations by location. Chennai demonstrates superior long-distance efficiency, maintaining 2-6 day delivery times even at maximum distances. In contrast, Delhi shows the most inconsistent performance with shipping times ranging from 2-9 days at similar distances, while Kolkata and Mumbai fall somewhere between these extremes. Bangalore performs well at medium distances but shows some variability at longer ranges.
## 📖 Story:
### This visualization tells the story of India's hub-and-spoke logistics network, where shipments likely originate from major distribution centers at specific distance intervals. The clustering of points suggests standardized shipping routes rather than random geographic distribution. Chennai emerges as the logistics champion, with its advanced infrastructure and strategic southern location enabling consistent performance regardless of distance. Delhi's erratic performance despite being a major hub indicates systemic challenges - possibly traffic congestion, regulatory bottlenecks, or inefficient last-mile delivery networks. # 3. HISTOGRAM - Shipping Times Distribution (Seaborn)

```python
fig = plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='shipping_times', bins=15, kde=True, alpha=0.7)
plt.title('Distribution of Shipping Times', fontsize=14, fontweight='bold')
plt.xlabel('Shipping Time (days)')
plt.ylabel('Frequency')
plt.tight_layout()

plt.tight_layout(pad=3.0)
plt.show()
```

### This histogram displays the frequency distribution of shipping times across all shipment records. It uses 15 bins to group the values, and includes a KDE (kernel density estimate) overlay to show the smooth probability curve.

## 🔍 Insight:
### Most shipments are delivered in a consistent time range (forming the central peak), but there’s a tail extending toward longer delivery times. The shape suggests a slightly right-skewed distribution, meaning while most deliveries are prompt, a few are delayed.

## 📖 Story:
### This plot tells us that the logistics system is generally efficient — most deliveries happen within a narrow time frame. However, the extended tail on the right indicates outliers: deliveries that take much longer. These delays, although rare, can have a big impact on customer satisfaction. By investigating and resolving the causes of these long deliveries, the company can improve reliability and build trust.

# 6. SCATTER PLOT - Products Sold vs Average Temperature

```python
fig = plt.figure(figsize=(20, 18))
plt.subplot(4, 2, 7)
disruption_colors = {0: 'green', 1: 'red'}
for disruption in [0, 1]:
    subset = df[df['disruption'] == disruption]
    plt.scatter(subset['avg_temp'], subset['number_of_products_sold'], 
               alpha=0.6, s=25, 
               color=disruption_colors[disruption],
               label=f'Disruption: {"Yes" if disruption else "No"}')
plt.title('Products Sold vs Average Temperature', fontsize=14, fontweight='bold')
plt.xlabel('Average Temperature (°C)')
plt.ylabel('Number of Products Sold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout(pad=3.0)
plt.show()
```

# 📈 Scatter Plot – Products Sold vs Average Temperature
## 📊 Description:

### This scatter plot shows the "Number of Products Sold" against "Average Temperature (°C)", with points colored by "Disruption" status (green for no, red for yes).
## 🔍 Insight:

### Product sales show a wide range across all observed temperatures (24-29°C), with no clear temperature-sales correlation. However, disruptions (red points) are notably concentrated around 28°C, suggesting this specific temperature may correlate with increased operational issues.
## 📖 Story:

### I wanted to see if temperature or disruptions affected sales. Sales vary widely regardless of temperature. Interestingly, disruptions frequently occur around 28°C. This pinpoints 28°C as a potential temperature to investigate for operational vulnerabilities, regardless of sales volume.

# 7. LINE PLOT - Average Shipping Time by Location

```python
fig = plt.figure(figsize=(12, 16))

plt.subplot(4, 2, 8)
location_avg = df.groupby('Location')['shipping_times'].mean().sort_values()
plt.plot(range(len(location_avg)), location_avg.values, marker='o', markersize=8, linewidth=2)
plt.title('Average Shipping Time by Location', fontsize=14, fontweight='bold')
plt.xlabel('Location')
plt.ylabel('Average Shipping Time (days)')
plt.xticks(range(len(location_avg)), location_avg.index, rotation=45)
plt.grid(True, alpha=0.3)

# Adjust layout to prevent overlap
plt.tight_layout(pad=3.0)
plt.show()
```

# 📈 Line Plot –Average Shipping Time by Location
## 📊 Description:

### This line plot visualizes the average shipping time in days across five different locations: Chennai, Bangalore, Kolkata, Mumbai, and Delhi. The y-axis represents average shipping time, and the x-axis represents the location.
## 🔍 Insight:

### Average shipping times vary by location, with Chennai and Bangalore having the shortest times (approx. 3.1 days), followed by Kolkata and Mumbai (approx. 3.5 days), and Delhi having the longest (approx. 3.7 days).
## 📖 Story:

### Our shipping data reveals a clear regional disparity. Chennai and Bangalore enjoy efficient deliveries, taking just over three days. However, as we move to Kolkata and Mumbai, average times increase by almost half a day. Delhi, our capital, experiences the longest delays, pushing nearly 3.7 days per shipment. This highlights critical areas for logistics improvement.
